```python
observed = pd.crosstab(train.gender, train.churn_encoded)
α=0.05
```

```python
chi2, p, _, expected = stats.chi2_contingency(observed)
if p < α:
    print('we can reject the null hypothesis')
else:
    print('we fail to reject the null hypothesis')
```
